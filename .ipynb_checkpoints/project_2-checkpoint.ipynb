## Preprocessing

```python
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
nltk.download('omw-1.4')
```

```python
def preprocessing(dataframe):
    
    processed_corpus = []
    stop_words = set(stopwords.words("english"))
    
    for i in tqdm(range(len(dataframe))):
        text = dataframe['Sentence'].iloc[i]
        
        # Remove tags
        text = BeautifulSoup(text).get_text()
        
        # Remove punctuations
        text = re.sub('[^a-zA-Z]', ' ', text)

        # Convert to lowercase
        text = text.lower()        
        
        # Convert to list from string
        text = text.split()

        # Lemmatization
        lem = WordNetLemmatizer()
        text = [lem.lemmatize(word) for word in text if not word in stop_words] 
        text = " ".join(text)
        processed_corpus.append(text)
        
    return processed_corpus
```

```python
processed_words = preprocessing(df)
dev_processed_words = preprocessing(dev_df)
```

```python
#processed_words
```

```python
df['Processed Words'] = processed_words
dev_df['Processed Words'] = dev_processed_words
```

### Bag of words

```python
bag_df = df.copy()
bag_dev_df = dev_df.copy()
```

```python
vectorizer = CountVectorizer()

bag = vectorizer.fit_transform(bag_df['Processed Words'])

#not sure about this. Using train dev bag to eval the dev set. Basically using training set features.
dev_bag = vectorizer.transform(bag_dev_df['Processed Words'])
```

```python
bag.toarray()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0, random_state=1, solver='lbfgs')
```

```python
lr.fit(bag.toarray(), bag_df['Feeling'])
```

```python
y_predict = lr.predict(dev_bag.toarray())
```

```python
from sklearn import metrics
```

```python
print("LogisticRegression Accuracy %.3f" 
%metrics.accuracy_score(bag_dev_df['Feeling'], y_predict))
```

```python
print("LogisticRegression F1 %.3f" 
%metrics.f1_score(bag_dev_df['Feeling'], y_predict, average = 'weighted'))
```

### TF-Idf (which is Similarity Metric, i guess)

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_df = df.copy()
tfidf_dev_df = dev_df.copy()
```

```python
tfidf_vectorizer = TfidfVectorizer()

tfidf_bag = tfidf_vectorizer.fit_transform(tfidf_df['Processed Words'])
dev_tfidf_bag = tfidf_vectorizer.transform(tfidf_dev_df['Processed Words'])



tfidf = TfidfTransformer()

tfidf_bag = tfidf.fit_transform(tfidf_bag, len(df['Processed Words']))
dev_tfidf_bag = tfidf.transform(dev_tfidf_bag, len(dev_df['Processed Words']))
```

```python
lr = LogisticRegression(C=100.0, random_state=1, solver='lbfgs')
lr.fit(tfidf_bag.toarray(), tfidf_df['Feeling'])
```

```python
y_predict = lr.predict(dev_tfidf_bag.toarray())
```

```python
print("LogisticRegression Accuracy %.3f" 
%metrics.accuracy_score(tfidf_dev_df['Feeling'], y_predict))
```

```python
print("LogisticRegression F1 %.3f" 
%metrics.f1_score(tfidf_dev_df['Feeling'], y_predict, average = 'weighted'))
```
